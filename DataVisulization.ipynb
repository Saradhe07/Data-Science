Matplotlib and Seaborn

```python
#importing libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({
     'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Sales': [100, 120, 90, 140, 160, 180, 150],
    'Customers': [30, 35, 25, 45, 50, 60, 40]
})
data
```

```python
#Matplotlib Basics
#Line plot
plt.plot(data['Day'],data['Sales'], marker='o', color='teal')
plt.title('Daily Sales Trend')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
#Bar plot
plt.bar(data['Day'],data['Customers'],color='Pink')
plt.title("Customers per Day")
plt.xlabel('Day')
plt.ylabel('Customers')
plt.show()
```

```python
#Pie Chart
plt.pie(data['Sales'], labels=data['Day'], autopct='%1.1f%%')
plt.title('Sales Distribution')
plt.show()
```

```python
#Histogram
marks = [45, 55, 67, 88, 76, 91, 60, 54, 80, 72, 50, 66]
plt.hist(marks, bins=5, color='yellow', edgecolor='black')
plt.title('Marks Distribution')
plt.xlabel('Marks Range')
plt.ylabel('Frequency')
plt.show()
```

Seaborn Basics



```python
#line plot
sns.lineplot(x='Day',y='Sales',data=data,marker='o')
plt.title('Sales Trend - Seaborn')
plt.show()
```

```python
# Bar Plot
sns.barplot(x='Day', y='Customers', data=data, palette='coolwarm')
plt.title('Customer Visits - Seaborn')
plt.show()
```

```python
#Histogram/Distribution
sns.histplot(marks, bins=5, kde=True, color='green')
plt.title('Marks Distribution - Seaborn')
plt.show()
```

```python
#Explore Relationships
#Goal: Find patterns between variables.
```

```python
# scatter plot
sns.scatterplot(x='Customers',y='Sales',data=data,s=100,color='pink')
plt.title('Sales vs Customers')
plt.show()
```

```python
#Regplot
sns.regplot(x='Customers',y='Sales',data=data,color='red')
plt.title('Regression Plot')
plt.show()
```

Explore Distributions & Correlations

```python
sns.boxplot(x='Sales',data=data, color='gold')
plt.title("Sales spred")
plt.show()
```

```python
sns.violinplot(x='Customers', data=data, color='lightblue')
plt.title('Customer Distribution Shape')
plt.show()
```

```python
# Heatmap (correlation)
sns.heatmap(data[['Sales', 'Customers']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
